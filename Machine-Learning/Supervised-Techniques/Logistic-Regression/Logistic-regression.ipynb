## Example Problem: Predicting Student Admission

### Problem Statement
We want to predict whether a student will be admitted to a university based on their exam score.

### Dataset
| Exam Score | Admission (1 = Yes, 0 = No) |
|------------|-----------------------------|
| 50         | 0                           |
| 60         | 0                           |
| 70         | 1                           |
| 80         | 1                           |
| 90         | 1                           |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

## Create the Data set

```python
# Dataset
exam_scores = np.array([50, 60, 70, 80, 90]).reshape(-1, 1)  # Feature (X)
admissions = np.array([0, 0, 1, 1, 1])  # Target (Y)
```

## Create the Logistic regression model

```python
model = LogisticRegression()
```

## Train the model

```python
model.fit(exam_scores, admissions)
```

# Run the model and make predictions

```python
predictions = model.predict(exam_scores)
predicted_probabilities = model.predict_proba(exam_scores)[:, 1]  # Probabilities for class 1
```

# Make predictions

```python
# Make predictions
predictions = model.predict(exam_scores)
predicted_probabilities = model.predict_proba(exam_scores)[:, 1]  # Probabilities for class 1
```

# Evaluate the model

```python
# Evaluate the model
accuracy = accuracy_score(admissions, predictions)
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(confusion_matrix(admissions, predictions))
print("Classification Report:")
print(classification_report(admissions, predictions))
```

## Visualize the result

```python
plt.scatter(exam_scores, admissions, color='blue', label='Actual Data')
plt.plot(exam_scores, predicted_probabilities, color='red', label='Sigmoid Curve')
plt.xlabel('Exam Score')
plt.ylabel('Probability of Admission')
plt.title('Logistic Regression: Admission Prediction')
plt.legend()
plt.show()
```

## Test the code with a random input

```python
# Example prediction for a new student
new_score = np.array([[75]])  # Example exam score
predicted_class = model.predict(new_score)
predicted_probability = model.predict_proba(new_score)[:, 1]
print(f"Predicted Class: {predicted_class[0]}")
print(f"Predicted Probability of Admission: {predicted_probability[0]:.2f}")
```
