## Example Problem: Predicting House Prices

### Problem Statement
We want to predict house prices based on the size of the house (in square feet).

### Dataset
| House Size (sqft) | Price ($) |
|--------------------|-----------|
| 800                | 150,000   |
| 1,000              | 200,000   |
| 1,200              | 240,000   |
| 1,400              | 280,000   |
| 1,600              | 310,000   |

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## Create the Data set

```python
# Dataset
house_size = np.array([800, 1000, 1200, 1400, 1600]).reshape(-1, 1)  # Feature (X)
house_price = np.array([150000, 200000, 240000, 280000, 310000])  # Target (Y)
```

## Create the Linear regression model

```python
model = LinearRegression()
```

## Train the model

```python
model.fit(house_size, house_price)
```

# Run the model and make predictions

```python
predicted_prices = model.predict(house_size)
```

## Calculate the Mean Squared Error

```python
mse = mean_squared_error(house_price, predicted_prices)
print(f"Mean Squared Error: {mse}")
```

## Visualize the result

```python
plt.scatter(house_size, house_price, color='blue', label='Actual Prices')
plt.plot(house_size, predicted_prices, color='red', label='Regression Line')
plt.xlabel('House Size (sqft)')
plt.ylabel('House Price ($)')
plt.title('House Price Prediction using Linear Regression')
plt.legend()
plt.show()
```

## Test the code with a random input

```python
# Example prediction for a new house
new_house_size = np.array([[1800]])  # Size of the new house
predicted_price = model.predict(new_house_size)
print(f"Predicted Price for 1800 sqft: ${predicted_price[0]:,.2f}")
```
